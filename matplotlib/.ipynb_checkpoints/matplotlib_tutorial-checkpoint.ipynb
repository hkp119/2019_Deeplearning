```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 1, 0.01)
print(x.shape)

# y = 2*x + epsilon, epsilon ~ n(0, 0.1)
y = np.random.randn(x.shape[0])
y = y * 0.1
y = y + 2 * x

plt.figure(figsize=(5, 5))
plt.plot(x, y, 'ro', label='raw data')
plt.suptitle('random datas')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.axis([0, 1, 0, 2.5])

plt.show()
```

```python
x2 = x * x
one = np.zeros(x.shape[0]) + 1

A = np.array([np.sum(x2), np.sum(x), np.sum(x), np.sum(one)])
A = A.reshape(2, 2)
Ainv = np.linalg.inv(A)
B = np.array([np.sum(x*y), np.sum(y)])
B = B.reshape(2, 1)
sol = np.dot(Ainv, B)
print(sol)

a = sol[0]
b = sol[1]

est = a * x + b

plt.plot(x, y, 'ro', label='raw data')
plt.plot(x, est, 'b--', label='estimated data')
plt.suptitle('LSE')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.axis([0, 1, 0, 2.5])

plt.show()
```

```python
plt.figure(figsize=(5, 7))
plt.subplot(211)
plt.plot(x, y, 'ro')

plt.subplot(212)
plt.plot(x, est, 'b-')
plt.show()
```

```python
# 이산적인 데이터 분류하기
class_name = np.array(['class A', 'class B', 'class C'])
class_value = np.array([34, 21, 68])

plt.figure(figsize=(10,5))
plt.subplot(131)
plt.bar(class_name, class_value)
plt.subplot(132)
plt.scatter(class_name, class_value)
plt.subplot(133)
plt.plot(class_name, class_value)

plt.suptitle('categorical plotting')
plt.show()
```

```python
# 히스토그램
uniform_dist = np.random.rand(500) + 40
gaussian_dist = np.random.randn(500) + 40

plt.figure(figsize=(10,10))
plt.subplot(211)
plt.hist(uniform_dist, bins=20)

plt.xlabel('parameter')
plt.ylabel('numbers')
plt.title('histogram example')
plt.grid(True)

plt.subplot(212)
plt.hist(gaussian_dist, bins=20)

plt.xlabel('parameter')
plt.ylabel('numbers')
plt.title('histogram example')
plt.grid(True)
plt.show()
```
