```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 1, 0.01)
print(x.shape)

# y = 2*x + epsilon, epsilon ~ n(0, 0.1)
y = np.random.randn(x.shape[0])
y = y * 0.1
y = y + 2 * x

plt.plot(x, y, 'ro', label='raw data')
plt.suptitle('random datas')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.axis([0, 1, 0, 2.5])

plt.show()
```

```python
x2 = x * x
one = np.zeros(x.shape[0]) + 1

A = np.array([np.sum(x2), np.sum(x), np.sum(x), np.sum(one)])
A = A.reshape(2, 2)
Ainv = np.linalg.inv(A)
B = np.array([np.sum(x*y), np.sum(y)])
B = B.reshape(2, 1)
sol = np.dot(Ainv, B)
print(sol)

a = sol[0]
b = sol[1]

est = a * x + b

plt.plot(x, y, 'ro', label='raw data')
plt.plot(x, est, 'b--', label='estimated data')
plt.suptitle('LSE')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.axis([0, 1, 0, 2.5])

plt.show()
```

```python
plt.figure()
plt.subplot(121)
plt.plot(x, y, 'ro')

plt.subplot(122)
plt.plot(x, est, 'b-')
plt.show()
```

```python
#   
class_name = ['class A', 'class B', 'class C']
class_value = [34, 21, 68]

plt.figure()
plt.subplot(131)
plt.bar(class_name, class_value)
plt.subplot(132)
plt.scatter(class_name, class_value)
plt.subplot(133)
plt.plot(class_name, class_value)

plt.suptitle('categorical plotting')
plt.show()
```
