```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-10,10,50)
y = np.linspace(-10,10,50)
X, Y = np.meshgrid(x, y)
```

```python
ep = np.random.randn(X.shape[0], X.shape[1])*10
z = 4.2 * X - 2.1 * Y + ep
z_est = 4.2 * X - 2.1 * Y
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X, Y, z)
ax.view_init(45, 30)
```

```python
def costfunction(x1, x2, X=X, Y=Y, z=z):
    ret = np.sum((x1*X + x2*Y - z) 
                * (x1*X + x2*Y - z))
    ret = 0.5 * ret
    ret = np.sqrt(ret)
    return ret

def gradient(f, x1, x2, X=X, Y=Y, z=z):
    h = 1e-4
    fx1h1 = f(x1+h, x2)
    fx1h2 = f(x1-h, x2)
    fx2h1 = f(x1, x2+h)
    fx2h2 = f(x1, x2-h)
    
    g1 = (fx1h1 - fx1h2) / (2*h)
    g2 = (fx2h1 - fx2h2) / (2*h)
    return g1, g2

def gd(f, x_init1, x_init2, iter_num=100, lr=0.01):
    x1 = x_init1
    x2 = x_init2
    for i in range(iter_num):
        ax.scatter(x1, x2, costfunction(x1, x2), s=40, c="r")
        g1, g2 = gradient(f, x1, x2)
        x1 -= lr*g1
        x2 -= lr*g2
    
    return x1, x2

beta1 = np.linspace(-5, 5, 50)
beta2 = np.linspace(-5, 5, 50)
B1, B2 = np.meshgrid(beta1, beta2)
cost = np.zeros(B1.shape)

for i in range(B1.shape[0]):
    for j in range(B2.shape[1]):
        cost[i, j] = (np.sum((B1[i, j]*X + B2[i,j]*Y - z) 
                    * (B1[i, j]*X + B2[i,j]*Y - z)))

cost = 0.5 * cost
cost = np.sqrt(cost)

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111, projection='3d')
ax.contour(B1, B2, cost, 40, corner_mask=True)
# ax.scatter(sol1, sol2, costfunction(sol1, sol2), s=120, c="r")

sol1, sol2 = gd(costfunction, 0, 0, lr=0.01, iter_num=50)
print(sol1)
print(sol2)
print(costfunction(sol1, sol2))

ep = np.random.randn(X.shape[0], X.shape[1])*10
z = 4.2 * X - 2.1 * Y + ep
print(costfunction(sol1, sol2, z=z))
```
